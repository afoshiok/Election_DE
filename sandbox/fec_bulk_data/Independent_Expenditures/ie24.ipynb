```python
import pandas as pd
import requests
import datetime
import zipfile
import os
import numpy as np
```

```python
zip_path = "./zipped/"
unzipped_path = "./unzipped/"

if not os.path.exists(zip_path):
    os.makedirs(zip_path)

if not os.path.exists(unzipped_path):
    os.makedirs(unzipped_path)
```

```python
ie_header_url = "https://www.fec.gov/files/bulk-downloads/data_dictionaries/pas2_header_file.csv"
ie_header_req = requests.get(ie_header_url)

ie_header_path = "./ie24_header.csv"

with open(ie_header_path, "wb") as ie_header_file:
    ie_header_file.write(ie_header_req.content)

print("Independent Expenditures header file successfully added!")
```

```python
ie_url = "https://www.fec.gov/files/bulk-downloads/2024/pas224.zip"
ie_req = requests.get(ie_url)

current_date = datetime.datetime.now()
current_date_fmt = current_date.strftime("%Y-%m-%d")

ie_output = f"./zipped/ie24_{current_date_fmt}.zip"

with open(ie_output, "wb") as ie_file:
    ie_file.write(ie_req.content)

print(f"Individual Expenditures file successfully added to this path: {ie_output}")
```

```python
ie_extract_path = f"./unzipped/ie24_{current_date_fmt}/"

with zipfile.ZipFile(ie_output) as extract_zip:
    extract_zip.extractall(ie_extract_path)

print(f"Files successfully extracted to this path: {ie_extract_path}")
```

```python
ie_header = pd.read_csv("./ie24_header.csv")
ie_df = pd.read_csv("./unzipped/ie24_2024-02-06/itpas2.txt", sep="|", names=ie_header.columns)

display(ie_df.head(n=20))
print(len(ie_df))
```

```python
print(ie_df.dtypes)
```

```python
ie_columns_change = ie_df[["ZIP_CODE","TRANSACTION_DT"]]
display(ie_columns_change.head(n=20))
```

```python
ie_df["TRANSACTION_DT"] = ie_df["TRANSACTION_DT"].fillna(0)
ie_df["TRANSACTION_DT"] = ie_df["TRANSACTION_DT"].astype(int)

ie_df["TRANSACTION_DT"] = pd.to_datetime(ie_df["TRANSACTION_DT"], format="%m%d%Y", errors='coerce')

display(ie_df["TRANSACTION_DT"].head(n=10))
```

```python
ie_df["ZIP_CODE"] = ie_df["ZIP_CODE"].astype(str)
# ie_df["ZIP_CODE"] = ie_df["ZIP_CODE"].str.split('.').str[0]
ie_df["ZIP_CODE"] = pd.to_numeric(ie_df["ZIP_CODE"], downcast="integer", errors='coerce').astype('Int64')
display(ie_df["ZIP_CODE"].head(n=20))
```

```python
display(ie_df.head(n=10))
print(ie_df.dtypes)
```

```python
ie_source_path = f"../source/ie24_{current_date_fmt}.csv"
ie_df.to_csv(ie_source_path, sep=",", index=False)

print(f"Source file for Independent Expenditures successfully added to: {ie_source_path}")
```
